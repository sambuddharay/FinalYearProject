```python
players_compiled = players_compiled.replace(np.nan, 0)
```

```python
X_data = players_compiled.drop(['Player','Nation', 'Pos', 'Squad', 'Born'], axis=1)
X_data
```

```python
kmeans_kwargs = {
    'init': 'random',
    'n_init': 50,
    'max_iter': 600,
    'random_state': 0
}
```

```python
pipeline = Pipeline([('scaler', MinMaxScaler()),
                     ('clustering', KMeans(n_clusters=4, **kmeans_kwargs))])

pipeline.fit(X_data)
```

```python
players.loc[:, "Cluster"] = pipeline[1].labels_
```

```python
players['Cluster']
```

```python
cluster_0 = players.loc[players['Cluster'] == 0]
cluster_1 = players.loc[players['Cluster'] == 1]
cluster_2 = players.loc[players['Cluster'] == 2]
cluster_3 = players.loc[players['Cluster'] == 3]
```
