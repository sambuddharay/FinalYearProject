```python
players = players.replace(np.nan, 0)
goalkeepers = goalkeepers.replace(np.nan, 0)
players_compiled = players.merge(goalkeepers, how='left')
players_compiled = players_compiled.replace(np.nan, 0)
```

```python
cat_info = players_compiled[['Player', 'Nation', 'Pos', 'Squad', 'Born']]
players_compiled = players_compiled.drop(columns=['Player', 'Nation', 'Pos', 'Squad', 'Born'])
```

```python
# Convert each row of the DataFrame into a NumPy matrix
player_matrices = [row.to_numpy() for index, row in players_compiled.iterrows()]

# Convert the list of matrices into a NumPy array
player_matrices_array = np.array(player_matrices)

player_matrices_array
```

```python
player_matrices_array.shape
```

```python
player_matrices_array = player_matrices_array.reshape(558,1,142)
player_matrices_array
```

```python
loss_df = pd.read_csv('https://raw.githubusercontent.com/sambuddharay/FinalYearProject/main/cnn/loss_data.csv')
```

```python
# Sort the DataFrame based on the absolute values of the 'Loss' column
df_sorted = loss_df.assign(Abs_Loss=loss_df['Loss'].abs()).sort_values(by='Abs_Loss')

# Extract the feature indexes of the 36 features with lowest loss
feature_indexes_lowest_loss = df_sorted.head(36)['Feature Index'].tolist()

# Print the feature indexes
feature_indexes_lowest_loss
```

```python
# Convert the list of feature indexes to NumPy array
feature_indexes_lowest_loss = np.array(feature_indexes_lowest_loss)

# Delete the selected features from X_train
X_train_filtered = np.delete(player_matrices_array, feature_indexes_lowest_loss, axis=2)

# Check the new shape of X_train_filtered
X_train_filtered.shape
```

```python
from keras.models import load_model

# Load the model
model = load_model("mymodel.keras")
```

```python
X_train_filtered = np.pad(X_train_filtered, ((0, 0), (0, 59), (0, 0)), mode='constant')
X_train_filtered.shape
```

```python
predictions = model.predict(X_train_filtered)
```

```python
predictions.shape
```

```python
df2 = pd.DataFrame(predictions, columns=['Score'])
player_score = pd.concat([cat_info, df2], axis=1)
player_score
```
